```python
df1_limit = df1[df1['Price_Type'] == 'Limit']
df1_limit.head()
```

```python
can.head()
```

```python
market = market.rename(columns = {'TimeStamp' : 'Timestamp'})
market.head()
```

```python
cancelled = df1_limit[
    df1_limit['Order Number'].isin(market['Order Number']) |
    df1_limit['Order Number'].isin(df1_stop['Order Number'])
].copy()

cancelled = cancelled.reset_index(drop=True)
cancelled.head(100000000000000)
```

```python
market[market['Order Number'] == '1300000000891934'].head()
```

```python
df1_stop[df1_stop['Order Number'] == '1300000000891934'].head()
```

```python
market_2 = market[market['Order Number'].isin(cancelled['Order Number'])]
stop_2 = df1_stop[df1_stop['Order Number'].isin(cancelled['Order Number'])]

print(len(market_2), len(stop_2))

```

```python
for _, o in cancelled.iterrows():
    time = int(o['Timestamp'])
    no = o['Order Number']
    can_time = 1e18

    ts_series = market_2.loc[market_2['Order Number'] == no, 'Timestamp']

    if not ts_series.empty and ts_series.notna().any():
        can_time = min(can_time, int(ts_series.min()))

    
    ts_series = stop_2.loc[stop_2['Order Number'] == no, 'Timestamp']
    if not ts_series.empty and ts_series.notna().any():
        can_time = min(can_time, int(ts_series.min()))

    if can_time == 1e18:
        print('alert')


    cancelled.loc[_, 'Timestamp'] = str(can_time)

    
    print(_, end='\r')

cancelled.head()

    

```
